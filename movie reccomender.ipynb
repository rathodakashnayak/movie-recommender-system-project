# Movie reccomder system 

```python
import pandas as pd
import numpy as np
```

```python
movie_data = pd.read_csv('movies.csv')
ratings_data = pd.read_csv('ratings.csv')
tags_data = pd.read_csv('tags.csv')
movies = pd.merge(movie_data, ratings_data, on='movieId')
movies = pd.merge(movies, tags_data, on=['movieId', 'userId'], how='left')
```

```python
movies.head()
```

```python
movies.tail(10)
```

```python
movies.info()
```

```python
movies = movies[['movieId','title','genres','rating','tag']]
```

```python
movies_unique = movies.drop_duplicates(subset='movieId')
movies_unique.head()
```

```python

movies_unique['rating'] = movies_unique['rating'].astype('int')
```

```python
movies_unique['genres']=movies_unique['genres'].str.replace('|', ',')
```

```python
movies_unique.head()
```

```python
movies_unique['genres'] = movies_unique['genres'].apply(lambda x: x.split())
```

```python
import re

# Remove year in parentheses from titles
movies_unique['title'] = movies_unique['title'].apply(lambda x: re.sub(r'\s*\(\d{4}\)', '', x)) 
```

```python
movies_unique['title'] = movies_unique['title'].apply(lambda x: x.lower())
```

```python
movies_unique.head()
```

```python
movies_unique['tags'] = movies_unique.apply(lambda row: row['genres'] + [str(row['rating'])]+[str(row['tag'])], axis=1)
```

```python
df = movies_unique[['movieId', 'title', 'tags']]
df.head(10)
```

```python
df = df.reset_index(drop=True) # Reset index after dropping duplicates
```

```python
df.tail(10) # Example of tags column
```

```python
df['tags'] = df['tags'].apply(lambda x: ' '.join(x)) # Convert list to string
```

```python
df['tags'][0] # Example of tags column
```

# Movie Recommender model building 

```python
# Movie Recommender model building
# vectorization
from sklearn.feature_extraction.text import CountVectorizer 
from sklearn.metrics.pairwise import cosine_similarity

# Vectorize the tags column
cv = CountVectorizer(max_features=5000, stop_words='english')
vector = cv.fit_transform(df['tags']).toarray()

# Calculate cosine similarity
similarity= cosine_similarity(vector)


```

```python
def recommend(movie):
    # Find all titles that contain the search string (case-insensitive)
    matches = df[df['title'].str.contains(movie.lower(), na=False)]
    if matches.empty:
        print(f"Movie '{movie}' not found. Here are some available titles:")
        print(df['title'].head(10).tolist())
        return
    movie_index = matches.index[0]
    distances = similarity[movie_index]
    movie_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:6]
    for i in movie_list:
        print(df.iloc[i[0]].title)
        
```

```python
print(df['title'].unique()[:20])  # Show first 20 titles
```

```python
recommend('batman')
```
